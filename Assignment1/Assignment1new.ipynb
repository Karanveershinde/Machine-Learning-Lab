```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data and rename columns
df = pd.read_csv("iris.csv")
df.columns = ['index', 's_width', 's_length', 'p_width', 'p_length', 'breed']

# 1. Histogram
df[['s_width', 's_length', 'p_width', 'p_length']].hist(bins=20, figsize=(10, 6), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Flower Measurements")
plt.show()

# 2. Bar chart
df['breed'].value_counts().plot(kind='bar', color='orange', edgecolor='black')
plt.title("Count of Each Breed")
plt.xlabel("Breed")
plt.ylabel("Count")
plt.show()

# 3. Pie chart
df['breed'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, shadow=True)
plt.title("Breed Distribution")
plt.ylabel("")
plt.show()

# 4. Box plot
sns.boxplot(x='breed', y='p_length', data=df)
plt.title("Petal Length by Breed")
plt.show()

# 5. Pair plot
sns.pairplot(df, hue='breed', vars=['s_width', 's_length', 'p_width', 'p_length'])
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()

# 6. Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['s_width', 's_length', 'p_width', 'p_length']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
